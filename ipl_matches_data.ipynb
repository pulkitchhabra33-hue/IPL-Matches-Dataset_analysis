```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('matches.csv')
df
```

```python
df= df.reset_index(drop= True)
df.index= df.index + 1
df
```

```python
df= df.drop(columns= 'method', axis= 1)
```

```python
df.info()
```

```python
df.drop(columns= 'id').describe()
```

```python
toss_decision_counts= df['toss_decision'].value_counts()
toss_decision_counts
```

```python
toss_winner_counts= df['toss_winner'].value_counts()
toss_winner_counts
```

```python
winner_counts= df['winner'].value_counts().reset_index()
winner_counts.columns= ['wins', 'team']
```

```python
sns.barplot(x= 'wins', y= 'team', data= winner_counts, color= 'skyblue')
plt.xticks(rotation= 90)
plt.title('IPL Winner Teams')
plt.xlabel('Teams')
plt.ylabel('Number of wins')
plt.show()
```

```python
top_player_of_match= df['player_of_match'].value_counts().head(10).reset_index()
top_player_of_match
```

```python
sns.barplot(x= 'player_of_match', y= 'count', data= top_player_of_match)
plt.title('Player of the IPL Matches')
plt.xlabel('Player_of_match')
plt.ylabel('Counts')
plt.xticks(rotation= 90)
plt.show()
```

```python
city_counts= df['city'].value_counts().head(10).reset_index()
city_counts
```

```python
sns.barplot(x='city', y='count', data= city_counts, color= 'orange')
plt.title('No. of cities of IPL Matches')
plt.xlabel('Cities')
plt.ylabel('Counts')
plt.xticks(rotation= 90)
plt.show()
```

```python
matches_per_season= df['season'].value_counts().sort_index().reset_index()
matches_per_season.index= range(1, len(matches_per_season)+1)
matches_per_season
```

```python
sns.barplot(x= 'season', y= 'count', data= matches_per_season, color= 'pink')
plt.title('IPL Matches per season')
plt.xlabel('Seasons')
plt.ylabel('No. of matches')
plt.xticks(rotation= 90)
plt.show()
```

```python
top_winners= df.groupby('season')['winner'].value_counts()
top_winners
```

```python
result_counts= df['result'].value_counts()
result_counts
```

```python
sns.barplot(x= result_counts.index, y= result_counts.values, color= 'orange')
plt.title('Match Result Distribution')
plt.xlabel('Result type')
plt.ylabel('No. of matches')
plt.xticks(rotation= 90)
plt.show()
```

```python
df['win_by_runs']= df.apply(lambda x: x['result_margin'] if x['result'] == 'runs' else 0, axis = 1)
df['win_by_runs']
```

```python
df['win_by_wickets'] = df.apply(lambda x: x['result_margin'] if x['result'] == 'wickets' else 0, axis=1)
df['win_by_wickets']
```

```python
#Winning data for the top 20 IPL teams
winning_data= df[['season','team1','team2','winner','result','result_margin','win_by_runs','win_by_wickets']].head(20)
winning_data.index= range(1, len(winning_data)+1)
winning_data
```

```python
df.nlargest(10, 'win_by_runs')[['team1', 'team2', 'winner', 'win_by_runs']].sort_index()
```

```python
df.nlargest(20, 'win_by_wickets')[['team1', 'team2', 'winner', 'win_by_wickets']].sort_values(by= 'win_by_wickets', ascending= False)
```

```python
#toss_decision plot
sns.countplot(x= 'toss_decision', data= df)
plt.title("Teams' toss_decision plot")
plt.show()
```

```python
df['margin']= df['win_by_runs'] + df['win_by_wickets']
df['margin']
```

```python
df['win_type']= df.apply(lambda x: 'Runs' if x['win_by_runs'] >0 else ('Wickets' if x['win_by_wickets'] >0 else None), axis= 1)
df
```
